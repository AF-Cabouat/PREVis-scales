```python
import os
import pandas as pd
import json

def process_json_folder(folder_path):
    data = {}

    # Iterate through files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith(".json"):
            file_path = os.path.join(folder_path, file_name)

            # Read the content of the .json file
            with open(file_path, 'r') as json_file:
                try:
                    json_data = json.load(json_file)
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON in {file_name}: {e}")
                    continue

                # Extract values with 3 digits after floating point
                formatted_data = {key: round(value, 3) for key, value in json_data.items()}

                # Add the data to the dictionary with file name as the key
                data[file_name[:-5]] = formatted_data

    # Create a DataFrame from the collected data
    df = pd.DataFrame(data).T

    return df

# Specify the folder path where your JSON files are stored
json_folder_path = '.'

# Process the folder and get the DataFrame
result_df = process_json_folder(json_folder_path)

result_df['average'] = result_df.mean(axis=1)

# Print the resulting DataFrame
result_df.to_csv('compare_greadability.csv')
result_df
```
